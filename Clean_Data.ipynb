```python
import pandas as pd
```

```python
data = pd.read_csv('Data/binhluanvienAnhQuan.csv')
```

```python
data
```

```python
data.info()
```

```python
# Loại bỏ cột NaN

data = data.dropna(axis=1, how='all')
```

```python
#Tách reactions
import ast

def split_reactions(row):
    try:
        reaction_dict = ast.literal_eval(row['reactions'])
        return pd.Series(reaction_dict)
    except (ValueError, SyntaxError):
        # Xử lý trường hợp không thể chuyển đổi thành từ điển
        return pd.Series()

reactions = data.apply(split_reactions, axis=1)
```

```python
reactions = reactions.fillna(0)
```

```python
reactions
```

```python
data['reactions'].dropna()
```

```python
data = pd.concat([data, reactions], axis=1)
data
```

```python
data['time'] = pd.to_datetime(data['time'])
data['time']
```

```python
data['hour'] = data['time'].dt.hour
data['day'] = data['time'].dt.day
data['month'] = data['time'].dt.month
data['year'] = data['time'].dt.year
data[['hour', 'day', 'month', 'year']]
```

```python
data.to_csv('Data/binhluanvienAnhQuan' + '_clean.csv', index=False)
```
